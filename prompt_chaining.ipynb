```python
from langgraph.graph import StateGraph , START , END
from langchain_ollama import ChatOllama
from typing import TypedDict
```

```python
model = ChatOllama(model="llama3.2:latest")
```

```python
class BlogState(TypedDict):
    topic : str
    outline : str
    blog_content : str
```

```python
def gen_outline(state:BlogState)->BlogState:
    topic = state['topic']
    prompt_outline = f"Generate a outline for the following {topic}"
    outline = model.invoke(prompt_outline).content
    state['outline'] = outline
    return state
```

```python
def gen_blog(state:BlogState)->BlogState:
    outline = state['outline']
    prompt_blog = f"Generate a detailed blog for the following {outline}"
    blog_content = model.invoke(prompt_blog).content
    state['blog_content'] = blog_content
    return state
```

```python
graph = StateGraph(BlogState)

graph.add_node("gen_outline",gen_outline)
graph.add_node("gen_blog",gen_blog)

graph.add_edge(START,"gen_outline")
graph.add_edge("gen_outline","gen_blog")
graph.add_edge("gen_blog",END)

workflow = graph.compile()
```

```python
output = workflow.invoke({'topic':"India Tourism"})
```
